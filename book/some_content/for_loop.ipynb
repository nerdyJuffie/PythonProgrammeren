# Nog meer herhaling ... for loops

### herhalen met de for-loop

Waar je bij de while loop deze moest onderbreken door te zorgen dat er òf een teller meeliep tot aan de verwachte waarde òf deze actief met een break moest afkappen, hoef je dit bij een for-loop niet te doen.

Een for-loop heeft als algemene syntax:

> \# x gaat van 0 TOT de eindwaarde  
> for x in range(eindwaarde):  
> &nbsp;&nbsp;&nbsp;&nbsp;doe iets

> \# x gaat van de beginwaarde TOT de eindwaarde  
> for x in range(beginwaarde, eindwaarde):  
> &nbsp;&nbsp;&nbsp;&nbsp;doe iets

> \# x gaat van de beginwaarde TOT de eindwaarde met stappen van stap  
> for x in range(beginwaarde, eindwaarde, stap):  
> &nbsp;&nbsp;&nbsp;&nbsp;doe iets
