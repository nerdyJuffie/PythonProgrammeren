# Verkorte Operatoren

## 1. Theorie: Verkorte schrijfwijze
In plaats van `aantal = aantal + 1` kun je in Python `aantal += 1` gebruiken.

| Operator | Betekenis | Voorbeeld | Lang uitgeschreven |
| :--- | :--- | :--- | :--- |
| `+=` | Optellen | `x += 3` | `x = x + 3` |
| `-=` | Aftrekken | `x -= 2` | `x = x - 2` |
| `*=` | Vermenigvuldigen | `x *= 5` | `x = x * 5` |
| `/=` | Delen | `x /= 2` | `x = x / 2` |
| `//=` | Floor Division | `x //= 3` | `x = x // 3` |
| `%=` | Modulo (Rest) | `x %= 4` | `x = x % 4` |
