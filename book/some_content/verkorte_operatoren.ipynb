#  Verkorte Operatoren

## 1. Theorie
Als we een waarde willen verhogen, kunnen we dat op twee manieren doen:

* **De lange manier:** `aantal_bananen = aantal_bananen + 1`
* **De korte manier:** `aantal_bananen += 1`

Dit werkt voor alle basisbewerkingen:
* `+=` (optellen)
* `-=` (aftrekken)
* `*=` (vermenigvuldigen)
* `/=` (delen)

````python
# Voorbeeld uit de uitleg: Bananen
aantal_bananen = 4
print("We hebben", aantal_bananen, "bananen")

aantal_bananen += 1 # Korte manier
print("Na aankoop hebben we er", aantal_bananen)

print("-" * 20)

# Jouw uitwerking van de Pinautomaat opdracht
print("Welkom bij de pinautomaat")
bedrag = int(input("Hoeveel euro wilt u opnemen? "))

vijftig = bedrag // 50
bedrag %= 50

twintig = bedrag // 20
bedrag %= 20

tien = bedrag // 10
bedrag %= 10

vijf = bedrag // 5
bedrag %= 5

twee = bedrag // 2
bedrag %= 2

een = bedrag // 1
bedrag %= 1

print("\nU krijgt:")
print(f"{vijftig}x 50, {twintig}x 20, {tien}x 10, {vijf}x 5, {twee}x 2, {een}x 1")
````
