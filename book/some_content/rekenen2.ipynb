# Rekenen in Python: De drie manieren van delen

In Python heb je drie verschillende operatoren om getallen door elkaar te delen. Elke operator geeft een ander resultaat.

### 1. De Enkele Deelstreep (`/`)
Dit is de normale deling. De uitkomst is een getal met decimalen (**float**).

### 2. De Dubbele Deelstreep (`//`)
Dit is de **gehele deling** (*floor division*). Het berekent hoe vaak een getal *volledig* in een ander getal past. De rest wordt weggegooid.

### 3. De Modulo (`%`)
De modulo berekent de **restwaarde** die overblijft na een gehele deling. Dit is erg handig om bijvoorbeeld te checken of een getal even of oneven is.

```python
# Voorbeeld: Snoepjes verdelen
snoepjes = int(input("Hoeveel snoepjes heb je? "))
kinderen = int(input("Hoeveel kinderen zijn er? "))

print("--- Resultaten ---")

per_kind_float = snoepjes / kinderen
print("Normale deling (/):", per_kind_float)

per_kind_geheel = snoepjes // kinderen
print("Hoeveel snoepjes krijgt ieder kind echt? (//):", per_kind_geheel)

overgebleven = snoepjes % kinderen
print("Hoeveel snoepjes blijven er over in de zak? (%):", overgebleven)
```
