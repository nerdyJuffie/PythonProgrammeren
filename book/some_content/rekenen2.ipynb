# Rekenen in Python: De drie manieren van delen

In deze les leer je hoe je getallen op verschillende manieren kunt delen met `/`, `//` en `%`.

### Uitleg
* **`/` (Normale deling)**: Geeft een getal met decimalen (bijv. `7 / 2 = 3.5`).
* **`//` (Gehele deling)**: Berekent hoe vaak een getal volledig past (bijv. `7 // 2 = 3`).
* **`%` (Modulo)**: Berekent wat er overblijft na de gehele deling (bijv. `7 % 2 = 1`).
