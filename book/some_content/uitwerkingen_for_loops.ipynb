### Oefening 1: De tafel van...
Dit programma vraagt om een getal en print de tafel van 1 tot 10 voor dat getal.

```python
tafel_van = int(input('Van welk getal wil je de tafel zien? '))
for teller in range(1, 11):
    print(f'{teller} * {tafel_van} = {teller * tafel_van}')
```

### Oefening 2: Aftellen maar...
Een klassieke countdown van 10 naar 0.

```python
for teller in range(10, -1, -1):
    print(teller)
    print("")
print("Happy New Year!!")
```

### Oefening 3: Tel alle even getallen op...
Dit programma berekent de som van alle even getallen *onder* het ingevoerde getal.

```python
getal = int(input("Tot welk getal moeten alle even getallen worden opgeteld? "))
som = 0
for teller in range(0, getal):
    if teller % 2 == 0:
        som += teller
print(f"De totale som is: {som}")
```
