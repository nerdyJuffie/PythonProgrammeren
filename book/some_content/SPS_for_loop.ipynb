```python
import random

score_speler = 0
score_computer = 0

for teller in range(0, 5):
    keuze = input("Steen, papier, schaar? ").lower()
    
    computerkeuze = random.randint(1, 3)
    if computerkeuze == 1:
        computer = 'steen'
    elif computerkeuze == 2:
        computer = 'papier'
    else:
        computer = "schaar"
    
    print(f"Computer kiest: {computer}")
    
    if keuze == computer:
        print("Gelijkspel")
    elif (keuze == 'steen' and computer == 'schaar') or \
         (keuze == 'papier' and computer == 'steen') or \
         (keuze == 'schaar' and computer == 'papier'):
        print("Punt voor jou!")
        score_speler += 1
    else:
        print("Punt voor de computer!")
        score_computer += 1
        
    print(f"Stand: {score_speler} - {score_computer}")
    print("-" * 20)

if score_speler > score_computer:
    print("Jij bent de uiteindelijke winnaar!")
elif score_speler < score_computer:
    print("De computer heeft het toernooi gewonnen.")
else:
    print("Het is een gelijkspel over 5 potjes!")
```
