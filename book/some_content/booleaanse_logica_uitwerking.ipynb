# Python Opdrachten: Booleaanse Logica

### Opdracht 1: De Uitsmijter

```python
leeftijd = int(input('Hoe oud ben je? '))
heeft_id = input('Heb je een ID bij je? (ja/nee) ')

mag_naar_binnen = leeftijd >= 18 and heeft_id == 'ja'

print('Toegang verleend:', mag_naar_binnen)
```

### Opdracht 2: De Kortingskaart

```python
leeftijd = int(input('Wat is de leeftijd? '))
korting = leeftijd < 12 or leeftijd > 65
print('korting:', korting)
```

### Opdracht 3: De Login-check

```python
gebruikersnaam = input('gebruikersnaam: ')
wachtwoord = input('wachtwoord: ')

login_geslaagd = gebruikersnaam == 'student_python' and wachtwoord == 'geheim'
print('login geslaagd?', login_geslaagd)
```

### Opdracht 4: De Complexere Check

```python
leeftijd = int(input('Wat is je leeftijd? '))
toestemming = input('Hebben je ouders toestemming gegeven? (ja/nee)') == 'ja'

toestemming_spel = (leeftijd >= 12 and toestemming) or leeftijd >= 16
print('Toegang tot spel:', toestemming_spel)
```
