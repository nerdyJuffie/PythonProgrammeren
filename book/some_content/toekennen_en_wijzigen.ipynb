# Waarden toekennen en wijzigen



## Een integer waarde (= geheel getal) toekennen aan een variabele

Er zijn twee manieren waarop je een integer waarde aan een variabele toe kan kennen


```python
getal = 13
print(getal)
# Met type( ) kun je het type van de variabele laten zien
print(type(getal))
```

2. Je kunt met _int( waarde of variabele)_ beloven aan de compiler dat de toegekende waarde of uitvoer een integer is


```python
getal = 16.8  # 16.8
print(type(getal))  # float
getal = int(getal)  # 16
print(type(getal))  # int
```

## Een floating point (float = kommagetal) waarde toekennen aan een variabele
1. Je kunt deze ook toekennen door het rechtstreeks toe te wijzen aan de variabele.

```python
float_getal = 14.6
print(float_getal)
print(type(float_getal))
```

2. Aan de compiler beloven dat de ingevoerde waarde een float is, dan wel een waarde omzetten naar een float
_float_getal = float(waarde of variabele)_

```python
float_getal = float(15)  # Hier wordt een integer omgezet naar een floating point getal
print(float_getal)
```

## Een string-waarde toekennen
1. Een tekst tussen aanhalingstekens (enkele of dubbele) toekennen: _tekst = 'Dit is een tekst'_

```python
tekst = 'Dit is een tekst'
print(tekst)
print(type(tekst))
```

## 7. Data Simulatie
Hier genereren we wat willekeurige getallen met `random`.

```python
import random
data = [random.randint(1, 100) for _ in range(5)]
print(f'Willekeurige getallen: {data}')
```
