# Waarden toekennen en wijzigen



## Een integer waarde (= geheel getal) toekennen aan een variabele

Er zijn twee manieren waarop je een integer waarde aan een variabele toe kan kennen
1. Door de integerwaarde rechtstreeks in te voeren. _variabelenaam = integerwaarde_


```python
getal = 13
print(getal)
# Met type( ) kun je het type van de variabele laten zien
print(type(getal))
```

2. Je kunt met _int( waarde of variabele)_ beloven aan de compiler dat de toegekende waarde of uitvoer een integer is


```python
a = 5
b = 10
som = a + b
print(f'De som van {a} en {b} is {som}')
```

## 3. Werken met Lijsten
Lijsten worden gebruikt om meerdere items in één variabele op te slaan.

```python
fruit = ['Appel', 'Banaan', 'Kers']
print(fruit)
```

## 4. For-loops
Met een `for` loop kun je door een lijst heen lopen.

```python
for item in fruit:
    print(f'Ik hou van {item}')
```

## 5. Functies Definiëren
Functies helpen om code herbruikbaar te maken.

```python
def begroeting(naam):
    return f'Hallo {naam}, welkom in dit notebook!'

print(begroeting('Gebruiker'))
```

## 6. Wiskundige Berekeningen
We kunnen formules ook beschrijven in Markdown, zoals $x^2 + y^2 = r^2$.

```python
import math
straal = 5
oppervlakte = math.pi * (straal ** 2)
print(f'Oppervlakte van de cirkel: {oppervlakte:.2f}')
```

## 7. Data Simulatie
Hier genereren we wat willekeurige getallen met `random`.

```python
import random
data = [random.randint(1, 100) for _ in range(5)]
print(f'Willekeurige getallen: {data}')
```
