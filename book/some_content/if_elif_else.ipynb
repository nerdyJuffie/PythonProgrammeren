# if... elif... else

Maar wat als je nu meerdere keuzes hebt, kun je dat dan wel oplossen in een python programma?

Jazeker, daarvoor hebben we de `elif`. De `elif` staat voor **anders als...**

### De syntax ziet er als volgt uit:

```python
if voorwaarde1:
    # als voorwaarde1 waar is wordt dit blok uitgevoerd
elif voorwaarde2:
    # als voorwaarde2 waar is wordt deze uitgevoerd
else:
    # als voorwaarde1 en voorwaarde2 niet waar zijn wordt dit blok uitgevoerd
```

### Voorbeeld: Kortingscategorieën
Stel je voor dat je korting krijgt als je binnen een bepaalde leeftijdscategorie valt of student bent, dan kun je dat mooi oplossen met een `if..elif.. else` codeblok.

````python
leeftijd = int(input("leeftijd "))
student = input("Ben je student? ")
korting = 0

if leeftijd < 10 or leeftijd > 65:
    korting = 10
elif student == "ja":
    korting = 5
else:
    korting = 0

print("Je hebt recht op", korting, "euro korting")
````

### Hoe werkt het?
Zoals je in het hierboven gegeven voorbeeld kunt zien, krijgt iemand van 9 die student is de volledige korting (10 euro). 

**De gouden regel:** Als aan de eerste voorwaarde in een `if..elif..else` blok voldaan wordt, dan wordt de rest **niet meer bekeken**. Dit geldt ook als aan de tweede voorwaarde voldaan wordt.

Je mag net zoveel `elif` voorwaarden gebruiken als je wilt.
