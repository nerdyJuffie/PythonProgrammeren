## 1. Willekeurige getallen in Python
In Python kun je met code willekeurige waarden laten maken. Hiervoor moet je de bibliotheek **random** importeren.

De algemene syntax is als volgt:
> `random.randrange(start, stop)`

* **start**: het eerste getal vanaf waar er gekozen kan worden.
* **stop**: tot welk getal er gekozen kan worden. Wil je dit getal er bij hebben, dan moet je als stop een getal hoger gaan zitten.

````python
import random

# Voorbeeld: random getal van 1 tot en met 10
willekeurigGetal = random.randrange(1, 11)
print(willekeurigGetal)

willekeurigGetal = random.randrange(1, 11)
print(willekeurigGetal)

willekeurigGetal = random.randrange(1, 11)
print(willekeurigGetal)
````

## 2. Opdracht: Steen, papier, schaar
Je gaat een klein spelletje Steen, Papier, Schaar maken, waarbij je speelt tegen de computer.

### Stappenplan:
1. Vertel de gebruiker dat deze een spel Steen, Papier, Schaar tegen de computer gaat spelen.
2. Vraag de gebruiker om een keuze te maken: **1 = Steen, 2 = Papier en 3 = Schaar**.
3. Maak een random waarde `keuzeComputer`. Deze kiest een random waarde uit tussen de 1 en 4. Je toont de keuze van de computer (met de naam erbij).
4. Vervolgens ga je de keuze van de speler en de computer vergelijken. **Tip:** maak een tabel om te kijken wie wanneer wint.
5. Met `if`, `elif`, `else` ga je (met hulp van je tabel) kijken wie er dit potje wint.



````python
# Start hier met je code voor Steen, Papier, Schaar
import random

````

```python
import random

# Steen Papier Schaar tegen de computer
print("je gaat zo steen, papier, schaar spelen tegen de computer")
keuze_speler = input("kies uit steen, papier of schaar: ")
keuze_c = random.randint(1,3)

if keuze_c == 1:
    keuze_computer = "steen"
elif keuze_c == 2:
    keuze_computer = "papier"
else:
    keuze_computer = "schaar"

print(f"Jij koos {keuze_speler} en de computer koos {keuze_computer}")

if keuze_speler == keuze_computer:
    print("gelijkspel")
elif (keuze_speler == "steen" and keuze_computer == "schaar") or \
     (keuze_speler == "papier" and keuze_computer == "steen") or \
     (keuze_speler == "schaar" and keuze_computer == "papier"):
    print("Jij wint van de computer")
else:
    print("De computer wint van jou!")
```
