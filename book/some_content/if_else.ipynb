# Het if-else commando

Soms wil je wat te kiezen hebben. Je kunt natuurlijk twee losse `if`-condities doen, maar je kunt deze ook combineren met een `else` (anders). 

Met `else` geef je aan wat het programma moet doen als de voorwaarde in de `if` **niet** waar is (**False**).

### Basis syntax
```python
if voorwaarde:
    codeblok dat uitgevoerd moet worden als voorwaarde waar is
else:
    codeblok dat uitgevoerd moet worden als voorwaarde onwaar is
```

### Voorbeeld: Stemgerechtigd
In dit voorbeeld controleren we of iemand oud genoeg is om te mogen stemmen.

````python
leeftijd = int(input("Hoe oud ben je? "))

if leeftijd >= 18:
    print("Je bent oud genoeg om te mogen stemmen")
else:
    print("Je zult nog moeten wachten tot je 18e jaar voordat je mag stemmen")
````
