# Het If-commando

Met het `if`-commando maak je beslissingen in je programma. De basis syntax voor een `if`-commando in Python is als volgt:

```python
if booleaanse expressie :
    code die uitgevoerd moet worden als de booleaanse expressie waar is.
```

### Inspringen (Indentation)
Je ziet dat de code die uitgevoerd moet worden als de booleaanse expressie waar is, **ingesprongen** is (meestal met 4 spaties of een Tab). Alle code die onder de `if` ingesprongen is, hoort bij het codeblok dat geldt als de expressie waar is.

Op het moment dat de code weer recht onder de `if` staat, wordt deze code **altijd** uitgevoerd, ongeacht de uitkomst van de test.

### Voorbeeld: Kinderkorting
Stel je wilt kijken of iemand kinderkorting kan krijgen, dan ga je testen of de leeftijd voldoet aan de voorwaarde (bijvoorbeeld 12 jaar of jonger).

````python
leeftijd = int(input("Wat is de leeftijd? "))
prijs = 20

if leeftijd <= 12:
    korting = 10
    prijs = prijs - korting

print("De toegangsprijs is", prijs)
````
