# Booleaanse Expressies en Logica

Een booleaanse expressie is een ‘test’ die de uitkomst **True** (waar) of **False** (onwaar) heeft.

### 1. Vergelijkingsoperatoren
Je maakt basisexpressies met deze operatoren:

| Operator | Betekenis |
| :--- | :--- |
| `==` | is gelijk aan |
| `!=` | is ongelijk aan |
| `<` | is kleiner dan |
| `>` | is groter dan |
| `<=` | is kleiner dan of gelijk aan |
| `>=` | is groter dan of gelijk aan |

```python
naam = "Sanne"
leeftijd = 23

stemmen = leeftijd > 18
kinderkorting = leeftijd <= 12

print(naam, leeftijd, "mag al stemmen:", stemmen)
print("Is de naam Sanne?", naam == "Sanne")
```

### 2. Combineren met `and` en `or` 

Als je meerdere booleaanse expressies tegelijkertijd wilt controleren, gebruik je `and` en/of `or`.

#### De `and` operator
Stel je wilt weten of iemand èn ouder is dan 17 jaar èn een rijbewijs heeft. Bij `and` is de uitkomst alleen **True** als **beide** kanten waar zijn.

```python
leeftijd = int(input("Wat is uw leeftijd? "))
rijbewijs = input("Heeft u uw rijbewijs? (ja/nee) ")

autorijden = leeftijd >= 17 and rijbewijs == "ja"

print("Mag al autorijden?", autorijden)
```

#### De `or` operator
Met `or` kun je kijken of er voldaan wordt aan òf de ene expressie òf de andere. Als aan **één van beiden** (of allebei) wordt voldaan, dan is de uitkomst **True**.

```python
leeftijd = int(input("Wat is uw leeftijd? "))

korting = leeftijd <= 10 or leeftijd > 65

print("Heb je recht op korting?", korting)
```
