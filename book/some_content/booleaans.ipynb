# Booleaanse expressies

Een booleaanse expressie is een ‘test’ die de uitkomst **True** (waar) of **False** (onwaar) heeft. 

Je maakt booleaanse expressies met vergelijkingsoperatoren. In Python heb je de volgende vergelijkingsoperatoren:

| Operator | Betekenis |
| :--- | :--- |
| `==` | is gelijk aan |
| `!=` | is ongelijk aan |
| `<` | is kleiner dan |
| `>` | is groter dan |
| `<=` | is kleiner dan of gelijk aan |
| `>=` | is groter dan of gelijk aan |

Met booleaanse expressies bepaal je hoe een programma verder moet gaan. Je kunt booleaanse expressies dus gebruiken om beslissingen in het programma te nemen.

### Oefening 1: Code overnemen
Voer de onderstaande cel uit om te zien hoe Python de vergelijkingen verwerkt.

```python
naam = "Sanne"
leeftijd = 23

stemmen = leeftijd > 18
kinderkorting = leeftijd <= 12

print(naam, leeftijd, "mag al stemmen:", stemmen)
print(naam, leeftijd, "krijgt kinderkorting:", kinderkorting)
print("Is de naam Sanne?", naam == "Sanne")
print("Is de naam Bas?", naam == "Bas")
print("Is 25-2 gelijk aan de leeftijd?", 25-2 == leeftijd)
```
