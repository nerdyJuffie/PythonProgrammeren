# Rekenen in Python

In Python kun je rekenen met `int` (gehele getallen) en `float` (kommagetallen). Hiervoor gebruik je de volgende **rekenkundige operatoren**:

| Operator | Bewerking | Voorbeeld |
| :--- | :--- | :--- |
| `+` | optelling | `5 + 2 = 7` |
| `-` | aftrekking | `5 - 2 = 3` |
| `*` | vermenigvuldiging | `5 * 2 = 10` |
| `/` | deling | `5 / 2 = 2.5` |
| `**` | machtsverheffing | `5 ** 2 = 25` |
| `%` | modulo (rest bepalen) | `5 % 2 = 1` |

### Rekenvolgorde
Bij het programmeren in Python (en andere programmeertalen) geldt dezelfde rekenvolgorde als bij wiskunde:
1. Haakjes `()`
2. Machtsverheffen `**`
3. Vermenigvuldigen `*` en Delen `/`
4. Optellen `+` en Aftrekken `-`

## Rekenen met variabelen
Variabelen gebruik je niet alleen om dingen op te slaan die je even later weer uitprint, je gebruikt variabelen ook om mee te rekenen. Dit maakt je code veel flexibeler.

```python
lengte = 10
breedte = 5
oppervlakte = lengte * breedte

print(f"De oppervlakte van een rechthoek van {lengte} bij {breedte} is: {oppervlakte}")
```
