# Herhalen met de while-loop

De eerste vorm van herhalen die we gaan behandelen is de while-loop. De while-loop heeft de volgende algemene vorm:

```python
teller = 0
while voorwaarde met teller:
    codeblok zolang voorwaarde geldt
    teller ophogen
```

---

```python
while voorwaarde:
    codeblok zolang voorwaarde geldt
    ( break om gegeven een bepaalde voorwaarde er uit te stappen)
```

### Voorbeelden met de teller:

````python
teller = 1

# het eerste getal is altijd het kleinste
kleinste = int(input("Voer een getal in: "))

# we gaan nu nog 4x om een getal vragen
# en we controleren of het getal het kleinste is
while teller < 5:
    getal = int(input("Voer een getal in: "))
    if getal < kleinste:
        kleinste = getal
        
    #verhoog de teller met 1
    teller += 1
    
print("Het kleinste getal is", kleinste)
````

````python
teller = 1
grens = int(input("Wat is de bovengrens van de teller? "))

while teller < grens:
    print("teller:", teller)
    if teller % 2 == 0:
        teller *= 2
    else:
        teller += 1
````

Dit programma telt binair totdat de bovengrens van de teller is bereikt. Wil je de waarde inclusief de grens hebben, dan had je `<=` moeten gebruiken in plaats van `<`.

Je kunt ook een voorwaarde zonder teller gebruiken en het programma laten stoppen met een break als er aan een bepaalde voorwaarde is voldaan

````python
teller = 1
grens = int(input("Wat is de bovengrens van de teller? "))
while teller < grens:
    if teller == 32:
        break
    print("teller:", teller)
    if teller % 2 == 0:
        teller *= 2
    else:
        teller += 1
````



In bovenstaand voorbeeld stopt het programma als de teller gelijk is aan de waarde 32. Als de teller 16 is, dan gaat het de loop in. 16 is te delen door 2, dus in regel 8 wordt de waarde van teller verdubbeld. De teller heeft nu waarde 32. De teller gaat opnieuw de loop in. Nu komt in regel 4 de test of de teller gelijk is aan 32. Dit is het geval en het programma stopt met de loop.
