# Opdracht
Je gaat het spel Steen-Papier-Schaar zo uitbreiden, dat je het spel blijft spelen totdat de speler aangeeft dat deze wil stoppen òf totdat een der spelers 5 punten heeft gescoord.
Je moet dus 2 globale variabelen `scoreSpeler` en `scoreComputer` maken, waarin je de punten bij houdt.

```python
import random

# Globale variabelen
scoreSpeler = 0
scoreComputer = 0

def start_spel():
    global scoreSpeler, scoreComputer
    opties = ["steen", "papier", "schaar"]
    
    print("Het spel begint! Typ 'stop' om te stoppen.")
    
    while scoreSpeler < 5 and scoreComputer < 5:
        print(f"\nStand - Jij: {scoreSpeler} | Computer: {scoreComputer}")
        
        keuze_speler = input("Maak je keuze (steen, papier, schaar): ").lower().strip()
        
        if keuze_speler == "stop":
            print("Je hebt het spel gestopt.")
            break
            
        if keuze_speler not in opties:
            print("Ongeldige keuze, probeer het opnieuw.")
            continue
            
        keuze_computer = random.choice(opties)
        print(f"Computer koos: {keuze_computer}")
        
        if keuze_speler == keuze_computer:
            print("Gelijkspel!")
        elif (keuze_speler == "steen" and keuze_computer == "schaar") or \
             (keuze_speler == "papier" and keuze_computer == "steen") or \
             (keuze_speler == "schaar" and keuze_computer == "papier"):
            print("Punt voor jou!")
            scoreSpeler += 1
        else:
            print("Punt voor de computer!")
            scoreComputer += 1

    print(f"\nEinduitslag: Jij {scoreSpeler} - {scoreComputer} Computer")

start_spel()
```

<details>
<summary><b>Klik hier voor de oplossing</b></summary>

De oplossing gebruikt een `while`-loop die controleert of beide scores lager zijn dan 5. 
Met de `global` declaratie kunnen we de variabelen `scoreSpeler` en `scoreComputer` binnen de functie aanpassen. 
De `break` zorgt ervoor dat het spel stopt zodra de gebruiker 'stop' typt.
