```python
import random

score_speler = 0
score_computer = 0

while score_speler < 5 and score_computer < 5:
    keuze = input("Steen, papier, schaar? (of 'stop'): ").lower()
    
    if keuze == 'stop':
        break
        
    opties = ['steen', 'papier', 'schaar']
    computer = random.choice(opties)
    print(f"Computer kiest: {computer}")
    
    if keuze == computer:
        print("Gelijkspel")
    elif (keuze == 'steen' and computer == 'schaar') or \
         (keuze == 'papier' and computer == 'steen') or \
         (keuze == 'schaar' and computer == 'papier'):
        print("Punt voor jou!")
        score_speler += 1
    else:
        print("Punt voor de computer!")
        score_computer += 1
        
    print(f"Stand: {score_speler} - {score_computer}")

if score_speler > score_computer:
    print("Jij bent de winnaar!")
else:
    print("De computer heeft gewonnen.")
```
