## 2. Werken met Variabelen
Pas de onderstaande variabelen aan en druk op 'Run'.

```python
naam = "Jouw Naam"
leeftijd = 16
print(f"{naam} is {leeftijd} jaar oud.")
```

## 3. Rekenen in Python
Python kan als rekenmachine worden gebruikt. Probeer een andere som!

```python
resultaat = 25 * 4 + 10
print(f"De uitkomst is: {resultaat}")
```

## 4. Lijsten aanmaken
Voeg zelf een extra item toe aan de lijst hieronder.

```python
kleuren = ["Rood", "Groen", "Blauw"]
print(kleuren)
```

## 5. Herhaling (For-loops)
Met een loop kun je acties herhalen.

```python
for kleur in kleuren:
    print(f"Ik zie de kleur {kleur}")
```

## 6. Voorwaarden (If-statements)
Pas de score aan om te zien wat er gebeurt.

```python
score = 75
if score >= 55:
    print("Voldoende!")
else:
    print("Helaas, onvoldoende.")
```

## 7. Functies gebruiken
Roep de functie aan met je eigen naam.

```python
def hallo(naam):
    return f"Hallo {naam}!"

print(hallo("Leerling"))
```
