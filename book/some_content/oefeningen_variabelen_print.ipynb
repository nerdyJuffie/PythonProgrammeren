## 2. Variabelen aanmaken
Een variabele is een soort 'doosje' waar je een waarde in stopt. Je geeft het doosje een naam zodat je het later terug kunt vinden.

```python
# Hier maken we een variabele 'school' aan
school = "Mijn School"
print(school)
```

```python
getal1 = 12
getal2 = 8
totaal = getal1 + getal2

print("De som is:")
print(totaal)
```
