## Input van de gebruiker

In de voorbeelden in de vorige paragraaf gaven wij als programmeurs waarden aan variabelen, maar we willen vooral programma's
maken die interactief zijn.
We kunnen programma's interactief maken door de gebruiker om __input__ te vragen.
Syntax: _variabelenaam = input('eventuele vraag of tekst')_

De variabele die via input() een waarde krijgt is altijd van het type _string_. Als je dus een integer, float of bool
wilt hebben, dan moet je de compiler beloven dat de input een int, float of bool is, door int( ), float( ) of bool( )

```python
naam = input('Wat is je naam? '
print('hallo' + naam)
leeftijd = input('hoe oud ben je?')
print(type(leeftijd))
```
