# Waarden toekennen via `input()`

Met de functie `input()` kun je aan de gebruiker via de console een vraag stellen. Het programma wacht dan tot de gebruiker een antwoord heeft getypt en op Enter heeft gedrukt. Pas daarna gaat het programma verder.

```python
variabele = input("hier komt jouw vraag aan de gebruiker")
```

````python
naam = input("Wat is jouw naam? ")
print(f"Hallo {naam}, leuk dat je er bent!")
````

## Het probleem: Alles is een String

De functie `input()` levert **altijd** een string (tekst) op. Dit is lastig als je aan je gebruiker een getal wilt vragen om mee te rekenen.

Kijk maar eens naar het voorbeeld hieronder. Als je 2 en 4 invoert, is de uitkomst niet 6, maar 24. Python heeft de twee teksten simpelweg aan elkaar geplakt. Dit noemen we **concateneren** of **concatenatie**.

````python
getal1 = input("Typ getal 2: ")
getal2 = input("Typ getal 4: ")

totaal = getal1 + getal2
print("De uitkomst van de optelling is:", totaal)
````

## De oplossing: Type Casting

We kunnen de `input()` functie 'inpakken' in een andere functie. Hiermee beloof je aan de **compiler** (het programma dat jouw code vertaalt naar computertaal) dat de invoer als een getal behandeld moet worden.

* Gebruik `int()` voor gehele getallen.
* Gebruik `float()` voor kommagetallen.

````python
# Nu pakken we de input in met int()
getal1 = int(input("Typ een geheel getal: "))
getal2 = int(input("Typ nog een geheel getal: "))

totaal = getal1 + getal2
print("Nu is de som wel correct:", totaal)
````
