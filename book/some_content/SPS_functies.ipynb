```python
import random

def bepaal_winnaar(speler, computer):
    # Deze functie krijgt de keuzes als parameters binnen
    if speler == computer:
        print("Gelijkspel")
        return 0
    elif (speler == 'steen' and computer == 'schaar') or \
         (speler == 'papier' and computer == 'steen') or \
         (speler == 'schaar' and computer == 'papier'):
        print("Punt voor jou!")
        return 1  # Geeft 1 punt terug voor de speler
    else:
        print("Punt voor de computer!")
        return 2  # Geeft een 'code' terug dat de computer wint

def speelSpel():
    score_speler = 0
    score_computer = 0

    for teller in range(0, 5):
        keuze = input("Steen, papier, schaar?  ").lower()
        
        computerkeuze = random.randint(1, 3)
        if computerkeuze == 1:
            computer = 'steen'
        elif computerkeuze == 2:
            computer = 'papier'
        else:
            computer = "schaar"
            
        print(f"Computer kiest: {computer}")

        # Hier roepen we de functie aan met de twee keuzes
        resultaat = bepaal_winnaar(keuze, computer)

        # De scores worden bijgewerkt op basis van wat de functie teruggeeft
        if resultaat == 1:
            score_speler += 1
        elif resultaat == 2:
            score_computer += 1
            
        print(f"Stand: {score_speler} - {score_computer}")

    # De einduitslag buiten de loop
    print("\n--- EINDSTAND ---")
    if score_speler > score_computer:
        print("Jij bent de winnaar!")
    elif score_speler < score_computer:
        print("De computer heeft gewonnen.")
    else:
        print("Het is gelijkspel!")

# Start het spel
speelSpel()
```
